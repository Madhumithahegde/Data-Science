```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Amazon Sales data.csv')

# Display the first few rows of the dataset and its structure
data.head(), data.info()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

# Extract year and month from 'Order Date'
data['Order Year'] = data['Order Date'].dt.year
data['Order Month'] = data['Order Date'].dt.month

data.head()
```

```python
import matplotlib.pyplot as plt

# Monthly Sales Trend
monthly_sales = data.groupby(['Order Year', 'Order Month'])['Total Revenue'].sum().reset_index()
monthly_sales_pivot = monthly_sales.pivot(index='Order Month', columns='Order Year', values='Total Revenue')
print(monthly_sales_pivot)

monthly_sales_pivot.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()

# Yearly Sales Trend
yearly_sales = data.groupby('Order Year')['Total Revenue'].sum().reset_index()

plt.bar(yearly_sales['Order Year'], yearly_sales['Total Revenue'])
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.show()
```

```python
import seaborn as sns

# Correlation matrix
correlation_matrix = data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Key Metrics')
plt.show()

# Scatter plot of Total Revenue vs Total Profit
sns.scatterplot(data=data, x='Total Revenue', y='Total Profit', hue='Item Type')
plt.title('
          ')
plt.show()
```
